# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`. You should've created these data files in the previous section: *Fixing Data Types Pt 3*.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline

# load datasets
df_08 = pd.read_csv("clean_08.csv")
df_18 = pd.read_csv("clean_18.csv")
```

```python
# Compare the distributions of greenhouse gas score in 2008 and 2018
df_08['greenhouse_gas_score'].plot(kind='hist', alpha=0.5);
df_18['greenhouse_gas_score'].plot(kind='hist', alpha=0.5);
plt.title('Distribution of Greenhouse Gas Score');
plt.xlabel('Greenhouse Gas Score');
labels = ['2008', '2018']
plt.legend(labels);
```

```python
# How has the distribution of combined mpg changed from 2008 to 2018
df_08['cmb_mpg'].plot(kind='hist', alpha = 1);
df_18['cmb_mpg'].plot(kind='hist', alpha = 0.5);
plt.title('Distribution of Combined MPG');
plt.xlabel('Combined MPG');
labels = ['2008', '2018']
plt.legend(labels);
```

```python
# Describe the correlation between displacement and combined mpg
ax=plt.gca()
ax.scatter(x=df_08['displ'], y=df_08['cmb_mpg'], color='tab:blue', alpha = 1);
ax.scatter(x=df_18['displ'], y=df_18['cmb_mpg'], color='tab:red', alpha = 0.5);
plt.title('Scatter PLot of Displacement vs Combined MPG');
plt.xlabel('Displacement');
plt.ylabel('Combined MPG');
plt.legend(['2008', '2018']);
```

```python
# correlation between greenhouse gas score and combined mpg
ax=plt.gca()
ax.scatter(x=df_08['greenhouse_gas_score'], y=df_08['cmb_mpg'], color='tab:blue', alpha = 1);
ax.scatter(x=df_18['greenhouse_gas_score'], y=df_18['cmb_mpg'], color='tab:red', alpha = 0.5);
plt.title('Scatter Plot of Greenhouse Gas Score vs Combined MPG');
plt.xlabel('Greenhouse Gas Score');
plt.ylabel('Combined MPG');
plt.legend(['2008', '2018']);
```
